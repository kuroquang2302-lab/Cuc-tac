```python
#bài tập tính thứ trong tuần của ngày 1/1/năm bất kỳ
import math
năm=int(input("Nhập số năm: "))
thứ=((năm+math.floor((năm-1)/4)-math.floor((năm-1)/100)+math.floor((năm-1)/400))%7)
x=int(năm)
y=int(thứ)
if y<6 and y>0:
    y="thứ "+str(y+1)
elif y==0:
    y="Chủ Nhật"
elif y==6:
    y="Thứ 7"

print("Năm:", x)
print("Ngày 1/1/",x, "là", y)
```
