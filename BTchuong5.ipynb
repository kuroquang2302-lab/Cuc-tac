```python
def computepay(a, b):
    if a>40:
        c=(40)*b + 1.5*(a-40)*b
    else:
        c=a*b
    return c

a=float(input("Enter Hours: "))
b=float(input("Enter Rate: "))

print("Enter Hours:", a)
print("Enter Rate:", b)
print("Pay", computepay(a,b))


```

```python
def computegrade(a):
    if a<0 or a>1:
        return("Bad score")
    elif a>=0.9:
        return("A")
    elif a>=0.8:
        return("B")
    elif a>=0.7:
        return("C")
    elif a>=0.6:
        return("D")
    else:
        return("F")

a=input("Enter Score: ")
try:
    a=float(a)
    print(f"Enter Score: {a}")
    print(computegrade(a))

except ValueError:
    print(f"Enter Score: {a}")
    print("Bad score")

```
