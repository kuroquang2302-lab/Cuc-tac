```python
#chương 3
#Exercise 1: Rewrite your pay computation to give the employee 1.5 times the hourly rate for hours worked above 40 hours.
print("Exercise 1")
H=float(input("Enter Hours: "))
R=float(input("Enter Rate: "))
if H>40:
    pay=40*R+(H-40)*R*1.5 #tính lương nếu làm trên 40h( lấy 40h nhân với lương cơ bản + số giờ làm thêm nhân với lương cơ bản x1.5)
else:
    pay=H*R
print("Enter Hours:", H)
print("Enter Rate:", R)
print("Pay:", pay)
```

```python
#Exercise 2: Rewrite your pay program using try and except so that your program handles 
# non-numeric input gracefully by printing a message and exiting the program.

H = input("Enter Hours: ")
try:
    H = float(H)
    # Chỉ khi H thành công, mới hỏi R
    R= input("Enter Rate: ")
    try:
        R = float(R)
        # Chỉ khi cả H và R thành công, mới tính toán
        if H > 40:
            pay = (40*R) + (H-40)*R*1.5
        else:
            pay = H * R
        print("Pay:", pay)
    except ValueError:
        print(f"Enter Hours: {H}")
        print(f"Enter Rate: {R}")
        print("Error, please enter numeric input")
except ValueError:
    print(f"Enter Hours: {H}")
    print("Error, please enter numeric input")
```

```python
#Exercise 6: Rewrite your pay computation with time-and-a-half for overtime and
#create a function called computepay which takes two parameters (hours and rate)

def computepay(H, R):
    if H>40:
        pay=40*R+(H-40)*R*1.5
    else:
        pay=H*R
    return pay

H=float(input("Enter Hours: "))
R=float(input("Enter Rate: "))
p=computepay(H, R)
print("Enter Hours:", H)
print("Enter Rate:", R)
print("Pay:", p)
```
